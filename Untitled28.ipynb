```python
# Задание task_03_02_01.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



def foo(a):                                                      #O(1)
    """Сортирует список методом пузырька (по возрастанию)

    Сложность: O(n**2) - квадратичная.
    """
    for i in range(len(a), 0, -1):                             #O(n)
        for j in range(1, i):                                  #O(n**2)
            if a[j-1] > a[j]:                                  #O(1)
                a[j-1], a[j] = a[j], a[j-1]                    #(1)
    return a                                                   #O(1)


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python

# Задание task_03_02_02.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



def foo(i):                                #O(1)
    """Преобразует целое число в строку, представляющую это число.


    Параметры:
        - i (int): число.

    Сложность: O(log n).
    """
    digits = "0123456789"                 #O(1)
    if i == 0:                            #O(1)
        return "0"                        #O(1)
    result = ""                           #O(1)
    while i > 0:                          #O(log n)
        result = digits[i%10] + result    #O(1)
        i = i // 10                       #O(1)
    return result                         #O(1)
```

```python
# Задание task_03_02_03.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



def foo(s):                      #O(1)
    """Суммирует числа в строке

    Параметры:
        - s (str): строка.

    Сложность: O(n).
    """
    val = 0                   #O(1)
    for c in s:               #O(n)
        if c.isdigit():       #O(1)
            val += int(c)     #O(1)
    return val               #O(1)
```

```python
# Задание task_03_02_04.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



def foo(n):                                 #O(1)
    """Возвращает список простых чисел от 1 до n.

    Параметры:
        - n (int): число.

    Сложность: O(n**2).
    """
    res = []                               #O(1)
    for i in range(1, n + 1):              #O(n)
        divisors = 0                       #O(1)
        j = 2                              #O(1)
        while j < i and divisors == 0:     #O(n)
            if i % j == 0:                 #O(1)
                divisors += 1              #O(1)
            j += 1                         #O(1)

        if divisors == 0:                  #O(n)
            res.append(i)                  #O(1)

    return res                             #O(1)
```

```python
# Задание task_03_02_08.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



import random
import time


def timeit(func, *args, **kw):
    """Выполнить функицю 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def maxmin1(data):                                      #O(1)
    """Вернуть минимум и максимум из 'data'.

    Алгоритм:
        Последовательный перебор всего массива.

    Сложность: O(n).

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    min_val = data[0]                                    #O(1)
    max_val = data[0]                                    #O(1)

    for num in data:                                     #O(n)
        if num < min_val:                                #O(1)
            min_val = num                                #O(1)
        if num > max_val:                                #O(1)
            max_val = num                                #O(1)

    return min_val, max_val                             #O(1)

def maxmin2(data):                                      #O(1)
    """Вернуть минимум и максимум из 'data'.

    Алгоритм:
        Сортировка списка, и возврат первого и последнего элементов.

    Сложность:  O(n log n) .

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    sorted_1 = sorted(data)                            #O(n log n)
    return sorted_1[0], sorted_1[-1]                   #O(1)


def maxmin3(data):                                   #O(1)
    """Вернуть минимум и максимум из 'data'.

    Алгоритм:
        Используя встроенные функции min() и max().

    Сложность: O(n).

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    return min(data), max(data)                     #O(n)


if __name__ == "__main__":                                                   #O(1)
    GEN_LIMIT = 1000000                                                      #O(1)
    dataset = []                                                             #O(1)
    for i in range(4, 7):                                                    #O(1)
        dataset.append(random.sample(range(-GEN_LIMIT, GEN_LIMIT), 10**i))   #O(1)

    res = [["i", "1 способ (мс.)", "2 способ (мс.)", "3 способ (мс.)"]]      #O(1)
    for data in dataset:                                                     #O(n)
        res.append([len(data),                                               #O(1)
                   timeit(maxmin1, data)[0],                                 #O(1)
                   timeit(maxmin2, data)[0],                                 #O(1)
                   timeit(maxmin3, data)[0]])                                #O(1)

    # Вывод
    print("{:>15} {:>15} {:>15} {:>15}".format(*res[0]))                     #O(1)
    for r in res[1:]:                                                        #O(n)
        print("{:>15} {:>15.2f} {:>15.2f} {:>15.2f}".format(*r))             #O(1)

# -------------
# Пример вывода:
#
#              i  1 способ (мс.)  2 способ (мс.)  3 способ (мс.)
#          10000            1.00            5.00            0.00
#            ...
```

```python
# Задание task_03_02_09.
#
# Выполнил: Войнова Екатерина Андреевна
# Группа: ЦИБ-241



import random
import time


def timeit(func, *args, **kw):
    """Выполнить функицю 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def is_ok_to_pass_1(db, passport_id):
    """Вернуть отметку о допуске из 'db' для паспорта с номером 'passport_id'.

    Если паспорт не найден в базе данных, возвращается отказ в допуске.

    Алгоритм:
        Последовательный перебор списка.

    Сложность: O(n)

    Параметры:
        - db (list of dict): база данных;
        - passport_id (str): номер паспорта.

    Результат:
        bool: отметка о допуске."""
    for result in db:                                       #O(n)
      if result['id'] == passport_id:                       #O(1)
            return result['ok']                             #O(1)

    return False                                            #O(1)


def is_ok_to_pass_2(db_opt, passport_id):
    """Вернуть отметку о допуске из 'db_opt' для паспорта с номером 'passport_id'.

    Если паспорт не найден в базе данных, возвращается отказ в допуске.

    Алгоритм:
        Поиск по ключу в словаре.

    Сложность: O(1)

    Параметры:
        - db_opt (dict): оптимизированная база данных для поиска;
        - passport_id (str): номер паспорта.

    Результат:
        bool: отметка о допуске."""
    return db_opt.get(passport_id, False)            #O(1)


def ok_str(flag):
    return "Да" if flag else "Нет"


if __name__ == "__main__":
    print("Загрузка базы данных... ")

    GEN_LIMIT = 1000000
    # База данных
    db = [{"id": "{:07d}".format(i), "ok": random.random() < 0.9}
          for i in random.sample(range(GEN_LIMIT, 10 * GEN_LIMIT), GEN_LIMIT)]

    # Структура db_opt должна быть оптимизированным хранилищем
    db_opt = {record['ok']: record['id'] for record in db}

    print("Загружено записей: {:,}.\n\nПримеры:".format(GEN_LIMIT))
    print("Первый элемент:", db[0])
    print("Последний элемент:", db[-1])
    print("\n")

    passport_id = input("Введите номер паспорта (7 цифр): ").strip()

    output = """
                  1 способ   2 способ
    Допущен     {:>10s} {:>10s}
    Время (мс.) {:>10.2f} {:>10.2f}\
    """

    res1 = timeit(is_ok_to_pass_1, db, passport_id)
    res2 = timeit(is_ok_to_pass_2, db_opt, passport_id)

    print(output.format(ok_str(res1[1]),
                        ok_str(res2[1]),
                        res1[0],
                        res2[0]))

# -------------
# Пример вывода:
#
# Загрузка базы данных...
# Загружено записей: 1,000,000.
#
# Примеры:
# Первый элемент: {'ok': True, 'id': '4998753'}
# Последний элемент: {'ok': True, 'id': '6280964'}
#
#
# Введите номер паспорта (7 цифр): 6280964
#
#               1 способ   2 способ
# Допущен             Да         Да
# Время (мс.)     145.10       0.00
```

```python
# Задание task_03_02_10.
#
# Выполнил: Войнова Екатерина
# Группа: ЦИБ 241



import random
import time


def timeit(func, *args, **kw):
    """Выполнить функицю 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def top3_1(db):
    """Вернуть список из ТОП-3 клиентов по сумме активов.

    Алгоритм:
        Сортировка (встроенный метод sort()) и возврат 3-х элементов.

    Сложность: O(n log n).

    Параметры:
        - db (list of dict): список клиентов.

    Результат:
        tuple of dict: 3 клиента с максимальной суммой активов."""
    sorted_db = sorted(db, key=lambda x: x['assets'], reverse=True)  #O(n log n)
    return tuple(sorted_db[:3])                                      #O(1)

def top3_2(db):
    """Вернуть список из ТОП-3 клиентов по сумме активов.

    Алгоритм:
        Прямой перебор (1 проход в цикле).

    Сложность: O(n).

    Параметры:
        - db (list of dict): список клиентов.

    Результат:
        tuple of dict: 3 клиента с максимальной суммой активов."""
    top_clients = []                                                    #O(1)

    for client in db:                                                  #O(n)
        if len(top_clients) < 3:                                       #O(1)
            top_clients.append(client)                                 #O(1)
            top_clients.sort(key=lambda x: x['assets'], reverse=True)  #O(1)
        else:                                                          #O(1)
            if client['assets'] > top_clients[-1]['assets']:           #O(1)
                top_clients[-1] = client                               #O(1)
                top_clients.sort(key=lambda x: x['assets'], reverse=True)#O(1)

    return tuple(top_clients)                                          #O(1)


if __name__ == "__main__":
    print("Загрузка базы данных... ")

    GEN_LIMIT = 10000000
    # Список клиентов
    db = []
    for client_id, assets in enumerate(
      random.sample(range(0, 10 * GEN_LIMIT), GEN_LIMIT),
      start=1):
        db.append(dict(client_id=client_id, assets=assets))

    print("Загружено записей: {:,}.\n\nПримеры:".format(GEN_LIMIT))
    print("Первый элемент:", db[0])
    print("Последний элемент:", db[-1])
    print("\n")

    res1 = timeit(top3_1, db)
    res2 = timeit(top3_2, db)

    for i, (func_time, res) in enumerate((res1, res2), start=1):
        print("Способ №{}".format(i))
        print("Время: {:.2f} мс.".format(func_time))
        for j, client in enumerate(res, start=1):
            print("{}-е место: ID={client_id} Активы={assets:,} руб.".
                  format(j, **client))
        print()


# -------------
# Пример вывода:
#
# Загрузка базы данных...
# Загружено записей: 10,000,000.
#
# Примеры:
# Первый элемент: {'client_id': 1, 'assets': 79149360}
# Последний элемент: {'client_id': 10000000, 'assets': 92745004}
#
#
# Способ №1
# Время: 16779.14 мс.
# 1-е место: ID=1346588 Активы=99,999,995 руб.
# 2-е место: ID=9924557 Активы=99,999,979 руб.
# 3-е место: ID=2904705 Активы=99,999,967 руб.
#
# Способ №2
# Время: 4364.90 мс.
# 1-е место: ID=1346588 Активы=99,999,995 руб.
# 2-е место: ID=9924557 Активы=99,999,979 руб.
# 3-е место: ID=2904705 Активы=99,999,967 руб.
```
