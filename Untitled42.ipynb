```python
import networkx as nx
import matplotlib.pyplot as plt
import time
from collections import deque

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

    def __repr__(self):
        return f"Node({self.val})"

class BSTVisualizer:
    def __init__(self):
        self.graph = nx.DiGraph()
        self.pos = {}
        self.fig, self.ax = plt.subplots(figsize=(10, 8))
        plt.ion()
        self.root_node = None

    def _get_tree_positions(self, node, x, y, level_width):
        if node is not None:
            self.pos[node.val] = (x, y)
            if node.left:
                self._get_tree_positions(node.left, x - level_width/2, y - 1, level_width/2)
            if node.right:
                self._get_tree_positions(node.right, x + level_width/2, y - 1, level_width/2)

    def update_graph(self, root, highlight_val=None, parent_val=None, inserted_val=None):
        self.ax.clear()
        self.graph.clear()

        if root is None:
            self.ax.set_title("Пустое дерево")
            self.fig.canvas.draw_idle()
            self.fig.canvas.flush_events()
            return

        nodes_to_process = deque([root])
        while nodes_to_process:
            node = nodes_to_process.popleft()
            self.graph.add_node(node.val)
            if node.left:
                self.graph.add_edge(node.val, node.left.val)
                nodes_to_process.append(node.left)
            if node.right:
                self.graph.add_edge(node.val, node.right.val)
                nodes_to_process.append(node.right)

        # Расположение узлов
        self._get_tree_positions(root, 0, 0, 4)

        node_colors = []
        for node_val in self.graph.nodes():
            if node_val == inserted_val:
                node_colors.append('lightgreen')
            elif node_val == highlight_val:
                node_colors.append('red')
            elif node_val == parent_val:
                node_colors.append('orange')
            elif node_val == getattr(self.root_node, 'val', None):
                node_colors.append('purple')
            else:
                node_colors.append('skyblue')

        nx.draw(self.graph, pos=self.pos, ax=self.ax,
                with_labels=True,
                node_color=node_colors,
                node_size=2000,
                font_size=10,
                font_weight='bold',
                arrowsize=20,
                edge_color='gray')

        self.ax.set_title("Бинарное дерево поиска")
        self.fig.canvas.draw_idle()
        self.fig.canvas.flush_events()

def bst_insert(root, val, visualizer=None, parent=None, delay=0.5):
    if root is None:
        new_node = TreeNode(val)
        if visualizer:
            visualizer.root_node = new_node
            visualizer.update_graph(visualizer.root_node)
            time.sleep(delay)
        return new_node

    if visualizer:
        visualizer.update_graph(visualizer.root_node if visualizer else root,
                                highlight_val=root.val,
                                parent_val=parent.val if parent else None)
        print(f"Текущий узел: {root.val}")
        time.sleep(delay/2)

    if val < root.val:
        root.left = bst_insert(root.left, val, visualizer, root, delay)
    elif val > root.val:
        root.right = bst_insert(root.right, val, visualizer, root, delay)
    else:
        print(f"Узел {val} уже существует")
        time.sleep(delay/2)

    if visualizer:
        visualizer.update_graph(visualizer.root_node)
    return root

def find_node(root, val, visualizer=None, parent=None , delay=0.5):
    if root is None:
        print(f"Узел {val} не найден")
        if visualizer:
            visualizer.update_graph(visualizer.root_node)
            time.sleep(delay)
        return None,None

    if visualizer:
        visualizer.update_graph(visualizer.root_node,
                                highlight_val=root.val,
                                parent_val=parent.val if parent else None)
        print(f"Поиск: проверяем узел {root.val}")
        time.sleep(delay)

    if val == root.val:
        print(f"Узел {val} найден!")
        time.sleep(delay)
        return root,parent
    elif val < root.val:
        return find_node(root.left,val ,visualizer,parent ,delay)
    else:
        return find_node(root.right,val ,visualizer,parent ,delay)

def find_min(node):
    parent = None
    current = node
    while current.left is not None:
        parent = current
        current = current.left
    return current,parent

def bst_delete(root,val ,visualizer=None ,delay=1):
    node,parent=find_node(root,val ,visualizer ,None ,delay)
    if node is None:
       return root

    # Случай: нет потомков или один потомок
    if node.left is None or node.right is None:
       new_child= node.left if node.left else node.right

       if parent is None: # удаляем корень
           root=new_child
       elif parent.left==node:
           parent.left=new_child
       else:
           parent.right=new_child

       if visualizer:
           visualizer.update_graph(visualizer.root_node ,inserted_val=val)
           print(f"Удален узел {val}")
           time.sleep(delay)

    # Случай: два потомка
    else:
       successor,succ_parent=find_min(node.right )
       # заменяем значение узла на значение преемника
       node.val=successor.val

       # удаляем преемника (у которого максимум в левом поддереве правого сына удаляемого узла)
       if succ_parent and succ_parent.left==successor :
           succ_parent.left=successor.right
       elif succ_parent and succ_parent.right==successor :
           succ_parent.right=successor.right

       # Обновляем граф после замены значения узла на преемника
       if visualizer :
           visualizer.update_graph(visualizer.root_node ,inserted_val=successor.val )
           print(f"Удален узел {val}, заменен на {successor.val}")
           time.sleep(delay)

    return root

def build_bst(elements ,visualizer ):
   if not elements :
       return None

   root=TreeNode(elements[0])
   visualizer.root_node=root
   visualizer.update_graph(root )
   time.sleep(1)

   for v in elements[1:] :
       root=bst_insert(root,v ,visualizer )
       visualizer.root_node=root

   return root

def heapify(arr,n,i):
   largest=i
   l=2*i+1
   r=2*i+2

   if l<n and arr[l]>arr[largest]:
       largest=l

   if r<n and arr[r]>arr[largest]:
       largest=r

   if largest!=i :
       arr[i],arr[largest]=arr[largest],arr[i]
       heapify(arr,n ,largest )

def heapsort(arr ):
   n=len(arr )

   # Построение max-heap
   for i in range(n//2 -1 ,-1 ,-1 ):
       heapify(arr,n,i )

   # Извлечение элементов из кучи
   for i in range(n-1 ,0 ,-1 ):
       arr[i],arr[0]=arr[0],arr[i]
       heapify(arr,i ,0 )

   return arr


def main():
    # Задача: Построить дерево из элементов {25;10;30;5;15;20;35}
    elements_bst = [25 ,10 ,30 ,5 ,15 ,20 ,35]

    print("Построение BST из элементов:", elements_bst)

    visualizer = BSTVisualizer()

    # Построение дерева и визуализация
    root_bst = build_bst(elements_bst ,visualizer )

    time.sleep(2)

    # Поиск элемента со значением 30
    print("\nПоиск элемента со значением 30:")
    find_node(root_bst ,30 ,visualizer )

    time.sleep(2)

    # Удаление элемента со значением 25 (корень или любой другой узел с этим значением)
    print("\nУдаление элемента со значением 25:")
    root_bst=bst_delete(root_bst ,25 ,visualizer )

    # Обновляем корень дерева после удаления (если он был удален или изменен)
    # В данном случае он может остаться тем же или стать новым корнем

    time.sleep(2)

    # Сортировка массива [25;10;30;5;15;20;35] с помощью heapsort
    array_to_sort=[25 ,10 ,30 ,5 ,15 ,20 ,35]

    print("\nСортировка массива с помощью Heapsort:")

     # Вывод исходного массива перед сортировкой
    print("Исходный массив:", array_to_sort )

    sorted_array=heapsort(array_to_sort.copy())

    print("Отсортированный массив:", sorted_array )


if __name__=="__main__":
     main()
```
